## EDA

### Question 1
For each country, what percentage of respondents reported a non-missing salary (CompTotal)?

What it asks: For each country, what share of respondents provided salary information?

Why it matters: Shows where compensation data is complete vs sparse, which helps decide whether to filter, weight, or impute countries when analyzing pay trends.

```python
%%bigquery --project=ba775-a01-477417
SELECT
  Country,
  COUNT(*) AS total_responses,
  COUNTIF(CompTotal IS NOT NULL) AS valid_salary_count,
  ROUND(COUNTIF(CompTotal IS NOT NULL)/COUNT(*)*100,2) AS valid_salary_percent
FROM `ba775-a01-477417.dataset.survey_2023_clean`
GROUP BY Country
ORDER BY valid_salary_percent DESC;
```

### Interpretation & Takeaway
This query shows, for each country, what share of respondents reported a non-missing salary (CompTotal). Some very small countries (e.g., Saint Kitts and Nevis, Monaco, Djibouti) have 100% completion but only 1–2 responses, so they don’t meaningfully inform pay trends. In major markets like the US, UK, and much of Western Europe, roughly 55–65% of respondents report salary, meaning compensation analysis is feasible there but still needs to account for substantial missingness when comparing countries.

### Question 2
How do technology preferences differ between new developers and experienced ones?

What it asks: Compares language and framework usage by experience level.

Why it matters: Reveals generational technology trends and shifting skills across the developer lifecycle.

```python
%%bigquery --project=ba775-a01-477417
SELECT Experience_Level, tech, tech_count
FROM (
  SELECT
    Experience_Level,
    tech,
    COUNT(*) AS tech_count,
    ROW_NUMBER() OVER (
      PARTITION BY Experience_Level
      ORDER BY COUNT(*) DESC
    ) AS rank
  FROM (
    SELECT
      CASE
        WHEN YearsCode < 3 THEN 'Beginner'
        WHEN YearsCode BETWEEN 3 AND 10 THEN 'Intermediate'
        ELSE 'Experienced'
      END AS Experience_Level,
      tech
    FROM `ba775-a01-477417.dataset.survey_2023_clean`,
    UNNEST(LanguageHaveWorkedWith) AS tech)
  GROUP BY Experience_Level, tech)
WHERE rank <= 5
ORDER BY
  CASE Experience_Level
    WHEN 'Beginner' THEN 1
    WHEN 'Intermediate' THEN 2
    WHEN 'Experienced' THEN 3
  END,
  rank;
```

### Interpretation & Takeaway
This query groups developers into Beginner, Intermediate, and Experienced and returns the top 5 languages they’ve worked with in each group. Across all three levels, JavaScript, HTML/CSS, Python, and SQL consistently appear at the top, showing a shared core stack regardless of experience. The main shift is that Java only appears for beginners, while TypeScript is prominent for intermediate and experienced developers, suggesting that more seasoned devs are more likely to adopt typed JavaScript.

### Question 3
Which tools or languages are developers most eager to learn, and how does that differ from what they already use?

What it asks: Compares current vs. desired technologies.

Why it matters: Identifies emerging tools and future skills in demand.

```python
%%bigquery --project=ba775-a01-477417
WITH current_usage AS (
  SELECT
    tech,
    COUNT(*) AS current_count
  FROM `ba775-a01-477417.dataset.survey_2023_clean`,
  UNNEST(LanguageHaveWorkedWith) AS tech
  GROUP BY tech),
desired_usage AS (
  SELECT
    tech,
    COUNT(*) AS desired_count
  FROM `ba775-a01-477417.dataset.survey_2023_clean`,
  UNNEST(LanguageWantToWorkWith) AS tech
  GROUP BY tech)

SELECT
  COALESCE(current_usage.tech, desired_usage.tech) AS Language,
  current_count,
  desired_count,
  ROUND(SAFE_DIVIDE(desired_count, current_count), 2) AS learn_to_use_ratio
FROM current_usage
FULL OUTER JOIN desired_usage
ON current_usage.tech = desired_usage.tech
ORDER BY learn_to_use_ratio DESC;
```

### Interpretation & Takeaway
This compares current vs. desired usage for each language and sorts by the learn_to_use_ratio (desired ÷ current). Languages like Zig, Rust, F#, Nim, Julia, and Elixir have very high ratios (2–5), meaning far more developers want to use them than currently do, marking them as strongly aspirational. In contrast, widely used staples like JavaScript, HTML/CSS, Python, Java, C#, SQL, and PHP have ratios well below 1, reflecting mature technologies that are common in practice but less exciting to adopt.

### Question 4
Do developers working with AI or data science technologies earn more than those in traditional software roles?

What it asks: Compares salaries by specialization.

Why it matters: Quantifies the financial return of AI-related skills.

```python
%%bigquery --project=ba775-a01-477417
WITH flat AS (
  SELECT
    TRIM(role) AS DevRole,
    CompTotal
  FROM `ba775-a01-477417.dataset.survey_2023_clean`,
  UNNEST(SPLIT(DevType, ';')) AS role
  WHERE CompTotal IS NOT NULL
        AND CompTotal > 0
        AND CompTotal < 1000000)

SELECT
  DevRole,
  ROUND(AVG(CompTotal), 2) AS avg_salary
FROM flat
GROUP BY DevRole
ORDER BY avg_salary DESC;
```

### Interpretation & Takeaway
These results clearly show a correlation between higher experience and a higher salary. For example a student's average salary is around $91,000, whereas a Developer Advocate is making over double. A developers experience across organizations and platforms is what leads to your success and learning several tools of the trade will help us earlier in our careers.


### Question 5
How does AI usage vary across different organization sizes?

What it asks: Compares micro (individual) and macro (organizational) adoption ratios.

Why it matters: Connects workforce readiness with business transformation pace.

```python
%%bigquery --project=ba775-a01-477417
WITH dev_ai AS (
  SELECT
    OrgSize,
    ROUND(
      SAFE_DIVIDE(
        COUNTIF(AIToolCurrentlyUsing NOT IN ('', 'I do not use AI tools', 'NA')),
        COUNT(*)
      ) * 100,2
    ) AS dev_ai_adoption_pct
  FROM `ba775-a01-477417.dataset.survey_2023_clean`
  GROUP BY OrgSize)

SELECT
  OrgSize,
  dev_ai_adoption_pct
FROM dev_ai
ORDER BY dev_ai_adoption_pct DESC;
```

### Interpretation & Takeaway
Smaller organizations (1–100 employees) are the most likely to use AI in their workflows, with over 40% of their work involving AI tools. In contrast, large companies with 5,000+ employees use AI for less than one-third of their work. This disparity may because of stricter AI policies and slower adoption processes in larger scale organizations.

### Question 6
How does developer confidence in AI tools relate to their years of coding experience?

What it asks: Tests whether seasoned developers are more or less confident in using AI tools than beginners.

Why it matters: Reveals if AI adoption confidence grows with experience or if newer developers adapt faster.


```python
%%bigquery --project=ba775-a01-477417
WITH exp_buckets AS (
  SELECT
    CASE
      WHEN SAFE_CAST(YearsCodePro AS INT64) BETWEEN 0 AND 2 THEN '0–2 years'
      WHEN SAFE_CAST(YearsCodePro AS INT64) BETWEEN 3 AND 5 THEN '3–5 years'
      WHEN SAFE_CAST(YearsCodePro AS INT64) BETWEEN 6 AND 10 THEN '6–10 years'
      WHEN SAFE_CAST(YearsCodePro AS INT64) BETWEEN 11 AND 20 THEN '11–20 years'
      WHEN SAFE_CAST(YearsCodePro AS INT64) > 20 THEN '20+ years'
      ELSE 'Unknown'
    END AS experience_level,

    CASE
      WHEN AIToolCurrentlyUsing IS NOT NULL
           AND AIToolCurrentlyUsing NOT IN ('', 'I do not use AI tools', 'NA')
      THEN 1 ELSE 0
    END AS uses_ai
  FROM `ba775-a01-477417.dataset.survey_2023_clean`
  WHERE YearsCodePro IS NOT NULL)

SELECT
  experience_level,
  ROUND(AVG(uses_ai) * 100, 2) AS ai_usage_rate
FROM exp_buckets
WHERE experience_level != 'Unknown'
GROUP BY experience_level
ORDER BY
  CASE experience_level
    WHEN '0–2 years' THEN 1
    WHEN '3–5 years' THEN 2
    WHEN '6–10 years' THEN 3
    WHEN '11–20 years' THEN 4
    WHEN '20+ years' THEN 5
    ELSE 6
  END;
```

### Interpretation & Takeaway
AI Usage is the most prominent for developers early in their career with 51% of developers 0-2 years in the field using the most. AI usage consistently goes down with more experience with only 28% of developers with over 20 years of experience using AI. This could be do to younger generations increased exposure to AI tools and beginning to implement them in the workforce. *italicized text*

What it asks: Identifies regions with strong AI engagement among individual developers.

Why it matters: Highlights leading geographies in AI-driven developer culture.

```python
%%bigquery --project=ba775-a01-477417
SELECT
  Country,
  COUNT(*) AS total_respondents,
  COUNTIF(AIToolCurrentlyUsing IS NOT NULL AND AIToolCurrentlyUsing NOT IN ('', 'I do not use AI tools', 'NA')) AS ai_users,
  ROUND(
    SAFE_DIVIDE(
      COUNTIF(AIToolCurrentlyUsing IS NOT NULL AND AIToolCurrentlyUsing NOT IN ('', 'I do not use AI tools', 'NA')),
      COUNT(*)
    ) * 100, 2
  ) AS ai_adoption_rate_pct
FROM `ba775-a01-477417.dataset.survey_2023_clean`
WHERE Country IS NOT NULL AND Country != ''
GROUP BY Country
-- Filter out countries with low sample size to ensure statistical significance
HAVING total_respondents > 100
ORDER BY ai_adoption_rate_pct DESC
LIMIT 15;
```

### Interpretation & Takeaway

Countries such as Tunisia (66.36%), Sri Lanka (62.67%), and Uruguay (60.90%) report the highest AI adoption rates among developers—meaning a majority of developers in these countries actively use AI tools. These regions show a strong AI-driven developer culture despite having smaller overall respondent counts.

This suggests that AI enthusiasm is growing fastest in emerging tech markets, where developers may rely on AI tools to compensate for limited resources, accelerate learning, or increase competitiveness in the global software economy.

## Question 10
Do developers in large organizations report more AI usage compared to those in smaller startups?

What it asks: Examines the relationship between the size of the organization (OrgSize) and the percentage of developers within those organizations who are actively using AI tools.

Why it matters: Helps students understand if joining a massive enterprise guarantees more exposure to advanced AI tools, or if smaller, more agile companies are actually leading the charge in daily adoption.

```python
%%bigquery --project=ba775-a01-477417
WITH dev_by_orgsize AS (
  SELECT
    OrgSize,
    AVG(
      CASE
        WHEN AIToolCurrentlyUsing IS NULL
             OR AIToolCurrentlyUsing IN ('', 'NA', 'I do not use AI tools')
        THEN 0
        ELSE 1
      END
    ) AS dev_ai_usage_rate
  FROM `ba775-a01-477417.dataset.survey_2023_clean`
  WHERE OrgSize IS NOT NULL
    AND OrgSize != ''
  GROUP BY OrgSize
),

org_overall_impact AS (
  SELECT
    COALESCE(Year, 2023) AS Year,
    AVG(Value) AS avg_headcount_increase_share
  FROM `ba775-a01-477417.dataset.state_of_ai_2023_clean`
  WHERE Exhibit = 'Impact of AI adoption on workforce'
    AND Category LIKE 'Increase by%'
  GROUP BY COALESCE(Year, 2023)
)

SELECT
  d.OrgSize,
  d.dev_ai_usage_rate,
  o.Year,
  o.avg_headcount_increase_share
FROM dev_by_orgsize d
CROSS JOIN org_overall_impact o
ORDER BY d.OrgSize, o.Year;
```

###Interpretation & Takeaway
The data indicates a clear inverse relationship between organization size and the adoption of AI tools by developers. Small organizations (2–19 employees) lead the charge with adoption rates reaching nearly 47%, likely driven by fewer bureaucratic hurdles and a greater need for efficiency. Conversely, developers in massive enterprises (10,000+ employees) report significantly lower usage at roughly 32%. This suggests that students seeking immediate, hands-on exposure to AI workflows might find more opportunities in startups and small businesses rather than large legacy corporations.

## Question 11
How does AI adoption vary by job function, and is high adoption associated with expected job stability?

What it asks: Groups developers by functional area (e.g., Product Development, Risk, Service Operations) to see who uses AI the most, and compares this with organizational expectations regarding headcount stability ("Expect no change").

Why it matters: Crucial for students choosing a specialization. It indicates which roles are embracing AI aggressively and whether that adoption is viewed by companies as a replacement for employees or a productivity enhancer.

```python
%%bigquery --project=ba775-a01-477417
WITH function_map AS (
  SELECT 'Product and/or service development' AS function_name, 'Back-end developer'                   AS devtype_pattern UNION ALL
  SELECT 'Product and/or service development',                              'Full-stack developer'                UNION ALL
  SELECT 'Product and/or service development',                              'Mobile developer'                    UNION ALL
  SELECT 'Service operations',                                              'DevOps'                              UNION ALL
  SELECT 'Marketing and sales',                                             'Front-end developer'                 UNION ALL
  SELECT 'Strategy and corporate finance',                                  'Engineering manager'                 UNION ALL
  SELECT 'Risk',                                                            'Security'                            UNION ALL
  SELECT 'Product and/or service development',                              'Data scientist or machine learning specialist'
),

dev_by_function AS (
  SELECT
    f.function_name,
    COUNT(*) AS dev_count,
    COUNTIF(
      s.AIToolCurrentlyUsing IS NOT NULL
      AND s.AIToolCurrentlyUsing NOT IN ('', 'I do not use AI tools', 'NA')
    ) AS ai_users,
    ROUND(
      100 * SAFE_DIVIDE(
        COUNTIF(
          s.AIToolCurrentlyUsing IS NOT NULL
          AND s.AIToolCurrentlyUsing NOT IN ('', 'I do not use AI tools', 'NA')
        ),
        COUNT(*)
      ),
      2
    ) AS dev_ai_adoption_pct
  FROM `ba775-a01-477417.dataset.survey_2023_clean` s
  JOIN function_map f
    ON s.DevType LIKE CONCAT('%', f.devtype_pattern, '%')
  GROUP BY f.function_name
),

org_by_function AS (
  SELECT
    Category AS function_name,
    SUM(CASE WHEN Metric LIKE 'Increase%' THEN Value ELSE 0 END) AS pct_expect_increase,
    SUM(CASE WHEN Metric LIKE 'Decrease%' THEN Value ELSE 0 END) AS pct_expect_decrease,
    SUM(CASE WHEN Metric LIKE 'Little or no change%' THEN Value ELSE 0 END) AS pct_expect_no_change
  FROM `ba775-a01-477417.dataset.state_of_ai_2023_clean`
  WHERE Exhibit = 'Effect of gen AI adoption on number of employees'
  GROUP BY Category
)

SELECT
  d.function_name,
  d.dev_count,
  d.dev_ai_adoption_pct,
  o.pct_expect_no_change
FROM dev_by_function d
LEFT JOIN org_by_function o USING (function_name)
ORDER BY d.function_name;
```

###Interpretation & Takeaway
When examining job functions, "Product and/or service development" emerges as the leader in AI adoption with a 54.35% usage rate, confirming that AI is now integral to core software building. There is a notable contrast between "Risk" and "Service Operations" roles; while Risk professionals have the lowest AI adoption (33.33%) and the highest expectation of headcount stability, Service Operations roles show higher AI usage but much lower expected stability. This implies that high AI adoption in operational roles may correlate with automation risks, whereas in core development, it acts more as a productivity enhancer.

## Question 12
Are individual developers adopting AI depth faster than organizations are scaling it?

What it asks: This query compares the "depth of adoption" at two levels. It calculates the percentage of individual developers who are "power users" (actively using AI specifically for search or development tasks) and compares it to the percentage of organizations that are "power adopters" (implementing AI across 3 or more distinct business functions).

Why it matters: This highlights the "maturity gap" between the workforce and the enterprise. It tells students whether they will likely be entering organizations that are already saturated with AI (top-down adoption) or if they will be the ones pushing for its use (bottom-up adoption).

```python
%%bigquery --project=ba775-a01-477417
WITH dev_power_users AS (
  SELECT
    ROUND(
      100 * AVG(
        CASE
          WHEN AIToolCurrentlyUsing IS NOT NULL
               AND AIToolCurrentlyUsing NOT IN ('', 'I do not use AI tools', 'NA')
               AND (
                 AISearchHaveWorkedWith IS NOT NULL OR
                 AIDevHaveWorkedWith IS NOT NULL
               )
          THEN 1
          ELSE 0
        END
      ),
      2
    ) AS power_user_share_pct
  FROM `ba775-a01-477417.dataset.survey_2023_clean`
),

org_multi_function AS (
  SELECT
    ROUND(AVG(Value), 2) AS multi_function_share_pct
  FROM `ba775-a01-477417.dataset.state_of_ai_2023_clean`
  WHERE Exhibit = 'Number of business functions adopting AI'
    AND Metric = '3 or more functions'
    AND (`Group` = 'All respondents' OR `Group` IS NULL)
)

SELECT
  d.power_user_share_pct,
  o.multi_function_share_pct
FROM dev_power_users d
CROSS JOIN org_multi_function o;
```

###Interpretation & Takeaway
The results reveal a significant lag between individual readiness and organizational scale. While 40.52% of developers classify as "power users"—deeply integrating AI into their core coding and search workflows—only 15.67% of organizations have managed to scale AI across three or more business functions. This suggests that while individuals have rapidly upskilled and integrated these tools, the broader enterprise machinery is slower to adapt. For students, this implies a career environment where they may possess more advanced AI workflows than the companies hiring them, positioning them as potential change agents who can help bridge this gap from the bottom up.
